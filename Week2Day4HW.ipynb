```python
from IPython.display import clear_output
class Pokemon:
    def __init__(self, pokename, poketype, level):
        self.pokename = pokename
        self.poketype = poketype
        self.level = level

        pass

    def display(self):
        print(f'Your {self.pokename} is a/an {self.poketype} type at Level {self.level}.')
        
class PocketList:
    def __init__(self):
        self.pocketlist = []
        
    def additem(self,item):
        self.pocketlist.append(item)
        
    
    def findpocketmon(self, pocketmon):
        for x in self.pocketlist:
            if pocketmon == x.pokename:
                return True
        return False
        
    def showinfo(self, pocketmon):
        for x in self.pocketlist:
            if pocketmon == x.pokename:
                {x.display()}

                

class Opponent:
    def __init__(self,oppname,weaknesses):
        self.oppname = oppname
        self.weaknesses = weaknesses
        pass
    
    def display(self):
        print(f'Your opponent will be {self.oppname}. Weakness: {self.weaknesses}.')
        pass
    
class OpponentList:
    def __init__(self):
        self.opplist = []
        
    def additem(self,item):
        self.opplist.append(item)
    
    def findopponent(self, rival):
        for x in self.opplist:
            if rival == x.oppname:
                return True
        return False
    
    def noopponent(self,rival):
        for x in self.opplist:
            if rival != x.oppname:
                return False
        return True
        
    def showinfo(self, rival):
        for x in self.opplist:
            if rival == x.oppname:
                {x.display()}

class Battlefield:
    def __init__(self, location):
        self.location = location
        pass
    
    def display(self):
        print(f'Your battlefield will be at {self.location}.')
        pass
    
class LocationList:
    def __init__(self):
        self.loclist = []
        
    def additem(self,item):
        self.loclist.append(item)
    
    def findlocation(self, local):
        for x in self.loclist:
            if local == x.location:
                return True
        return False
        
    def showinfo(self, local):
        for x in self.loclist:
            if local == x.location:
                {x.display()}      


class Main:
    @classmethod
    def run(self):
        print(f"Welcome to the PokeBattle Selection Assistant!")
        Pikachu = Pokemon('Pikachu','Electric',35)
        Charmander = Pokemon('Charmander','Fire',34)
        Bulbasaur = Pokemon("Bulbasaur","Grass",35)
        Squirtle = Pokemon("Squirtle","Water",56)
        Togepi = Pokemon("Togepi","Fairy",35)
        Magikarp = Pokemon("Magikarp","Water",50)
        Metapod = Pokemon("Metapod","Bug",100)
        Jigglypuff = Pokemon("Jigglypuff","Fairy",10)
        Diglett = Pokemon("Diglett","Ground",58)
        Psyduck = Pokemon("Psyduck","Water",45)
        Vaporeon = Pokemon("Vaporeon","Water",25)
        Koffing = Pokemon("Koffing","Poison",86)
        Dragonite = Pokemon("Dragonite","Dragon",2)
        
        poclist = PocketList()
        poclist.additem(Pikachu)
        poclist.additem(Charmander)
        poclist.additem(Bulbasaur)
        poclist.additem(Togepi)
        poclist.additem(Magikarp)
        poclist.additem(Metapod)
        poclist.additem(Jigglypuff)
        poclist.additem(Diglett)
        poclist.additem(Psyduck)
        poclist.additem(Vaporeon)
        poclist.additem(Koffing)
        poclist.additem(Squirtle)
        poclist.additem(Dragonite)

        Ash = Opponent("Ash","lacks maturity/has mommy issues")
        Misty = Opponent("Misty","literally hates Psyduck")
        Brock = Opponent("Brock","Nurse Joy or Officer Jenny")
        Jessie = Opponent("Jessie", "James")
        James = Opponent("James","Jessie")
        opplist = OpponentList()
        opplist.additem(Misty)
        opplist.additem(Brock)
        opplist.additem(Jessie)
        opplist.additem(James)
        opplist.additem(Ash)
        
        Vic = Battlefield("Victory Road")
        Ceru = Battlefield("Cerulean City")
        Pewter = Battlefield("Pewter City")
        loclist = LocationList()
        loclist.additem(Vic)
        loclist.additem(Ceru)
        loclist.additem(Pewter)
        
        while True:
            choose_opp = input("Please choose your Opponent (Ash/Misty/Brock/Jessie/James) or say 'exit'")
            if opplist.findopponent(choose_opp.title()):
                opplist.showinfo(choose_opp.title())
                break
            elif choose_opp.lower() == 'exit':
                print(f'Ripperoni Pepperoni :(')
                break
            elif choose_opp != opplist:
                print(f'Sorry, you have not yet met that Trainer, please make more trouble.')
                continue

       
        while True:  
            if choose_opp.lower() == 'exit':
                break
            choose_pokemon1 = input("Please choose your first pokemon or say 'exit'")
            if poclist.findpocketmon(choose_pokemon1.title()):
                poclist.showinfo(choose_pokemon1.title())
                break
            elif choose_pokemon1.lower() == 'exit':
                print(f'Ripperoni Pepperoni :(')
                break
            elif choose_pokemon1 != poclist:
                print(f'Sorry, you have not yet caught that Pokemon! Please choose another one.')
                continue
       
                
        while True:    
            if choose_opp.lower() == 'exit':
                break 
            elif choose_pokemon1.lower() == 'exit':
                break
            choose_pokemon2 = input("Please choose your second pokemon or say 'exit'")
            if choose_pokemon2.lower() == 'exit':
                print(f'Ripperoni Pepperoni :(')
                break
            elif choose_pokemon2 == choose_pokemon1:
                print(f'Nah b, pick a different one bro')
                continue
            elif poclist.findpocketmon(choose_pokemon2.title()):
                poclist.showinfo(choose_pokemon2.title())
                break
            elif choose_pokemon2 != poclist:
                print(f'Sorry, you have not yet caught that Pokemon! Please choose another one.')
                continue

                
        while True: 
            if choose_opp.lower() == 'exit':
                break 
            elif choose_pokemon1.lower() == 'exit':
                break
            elif choose_pokemon2.lower() == 'exit':
                break
            choose_pokemon3 = input("Please choose your final pokemon or say 'exit'")
            if choose_pokemon3.lower() == 'exit':
                print(f'Ripperoni Pepperoni :(')
                break            
            elif choose_pokemon3 == choose_pokemon1:
                print(f'Nah b, pick a different one bro')
                continue
            elif choose_pokemon3 == choose_pokemon2:
                print(f'Nah b, pick a different one bro')
                continue
            elif poclist.findpocketmon(choose_pokemon3.title()):
                poclist.showinfo(choose_pokemon3.title())
                break
            elif choose_pokemon3 != poclist:
                print(f'Sorry, you have not yet caught that Pokemon! Please choose another one.')
                continue

        while True:
            if choose_opp.lower() == 'exit':
                break  
            elif choose_pokemon1.lower() == 'exit':
                break
            elif choose_pokemon2.lower() == 'exit':
                break
            elif choose_pokemon3.lower() == 'exit':
                break 
            choose_loc = input("Please choose your Battle Location (Victory Road/Cerulean City/Pewter City) or say 'exit'")
            if choose_loc.lower() == 'exit':
                print(f'Ripperoni Pepperoni :(')
                break
            elif loclist.findlocation(choose_loc.title()):
                loclist.showinfo(choose_loc.title())
                break
            elif choose_pokemon3 != poclist:
                print(f'Sorry! You have not yet discovered this area, please explore more.')
                continue
                
        while True:    
            if choose_opp.lower() == 'exit':
                break  
            elif choose_pokemon1.lower() == 'exit':
                break
            elif choose_pokemon2.lower() == 'exit':
                break
            elif choose_pokemon3.lower() == 'exit':
                break 
            elif choose_loc.lower() == 'exit':
                break
            editpoke = input("Would you like to restart your Battle Selections or confirm your Battle or QUIT? Please choose restart/confirm/quit")
            if editpoke == 'quit':
                print(f'You have been removed your Battle Information. GG')
                break
                
            elif editpoke == 'confirm':
                clear_output()
                print(f"Your next battle has been confirmed.")
                print(f"You will be battling {choose_opp.title()} at {choose_loc.title()} with your 3 chosen Pokemon: {choose_pokemon1.title()}, {choose_pokemon2.title()}, & {choose_pokemon3.title()}. Good luck!!")
                break
            elif editpoke == 'restart':
                clear_output()
                Main.run()
        while True:    
            if choose_opp.lower() == 'exit':
                break  
            elif choose_pokemon1.lower() == 'exit':
                break
            elif choose_pokemon2.lower() == 'exit':
                break
            elif choose_pokemon3.lower() == 'exit':
                break 
            elif choose_loc.lower() == 'exit':
                break 
            another_one = input("Would you like to set up another battle? Yes/No")
            if another_one.lower() == "yes":
                clear_output()
                print(f"Your previous battle was against {choose_opp.title()} at {choose_loc.title()} with your 3 chosen Pokemon: {choose_pokemon1.title()}, {choose_pokemon2.title()}, & {choose_pokemon3.title()}.")
                Main.run()
            elif another_one.lower() == "no":
                print(f'Until next time on another episode of... POKEMON! Gotta Catch Em All!')
                break

                

```
